```python
import sqlite3
import pandas as pd
from datetime import datetime, UTC

# Load the data from staging table
conn = sqlite3.connect("boxing_odds_staging.db")
stg_df = pd.read_sql("SELECT * FROM raw_odds", conn)

# Add observed timestamp with timezone-aware UTC
stg_df["observed_at"] = datetime.now(UTC).strftime('%Y-%m-%d %H:%M:%S')

# Ensure the 'odds_history' table exists with matching columns before this step
# Append to odds_history table
stg_df.to_sql("odds_history", conn, if_exists="append", index=False)

conn.close()
```
