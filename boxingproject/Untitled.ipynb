```python
import sqlite3
conn = sqlite3.connect("boxing_odds_staging.db")
print("Tables:")
for (t,) in conn.execute("SELECT name FROM sqlite_master WHERE type='table'"):
    print("-", t)
print("\nDDL:")
for (sql,) in conn.execute("SELECT sql FROM sqlite_master WHERE type!='meta' AND sql NOT NULL"):
    print(sql, "\n")
conn.close()
```
